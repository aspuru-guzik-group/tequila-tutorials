# Universal Perturbative SF-[2]$_{R12}$ Correction

```python
import tequila as tq
import numpy
import psi4
```

In this tutorial (based on [1]), we present the integration of the (spin-free) universal F12 correction by [5], initially proposed in a
spin-orbital form in [4]. This perturbative correction takes as input only the 1- and 2-RDM (see [ReducedDensityMatrices-tutorial](https://github.com/aspuru-guzik-group/tequila/blob/master/tutorials/ReducedDensityMatrices.ipynb))
 of an arbitrary reference method. This correction aims to account for the basis-set
incompleteness due to the correlation cusp, and aims to fish in the same waters as methods based on the quantum
subspace expansion (QSE). Recently, a similar approach, yet as an a priori modification, has been carried out by
[7] and [8], incorporating the transcorrelated Hamiltonian as explicitly correlated method. The difference is here, that
the RDMs do not need to be resolved, but additional integrals need to be computed beforehand. Looking only at the additional cost on
classical computers, they both scale approximately $\mathcal{O}(N^6)$ in the particle number (although the version currently implemented in `tequila` scales as $\mathcal{O}(N^8)$). An integration of `[2]-R12`
into VQE makes insofar sense, because often, the 1- and 2-RDM need to be measured for the evaluation of general observables,
and maybe one wants to leave the input-molecular integrals as is.

To run this tutorial, you need to install `Psi4` from [this repository](https://github.com/philipp-q/psi4/tree/ri_space) for the GBS-based calculations. Molecular integrals for the MRA-PNOs are provided.

## Procedure
The overall procedure goes as follows:
1. Compute a reference. Here, we pick VQE. 
1. Set up the correction. This broadly needs the 1- and 2-RDM, an additional CABS basis and a f12-exponent as input.
    - Compute the RDMs. 
        - Precompute the RDMs and hand over explicitly + code ex
        - Specify a quantum circuit and a set of (fixed) parameters
        - Specify a dictionary specifying instruction to compute the RDMs via CI in `Psi4` 
    - The CABS basis is necessary for a _resolution of the identity_ approximation, that is used to avoid higher-order molecular integrals, and serves as well as additional space beyond the reference, that the perturbative method exploits in order to account for the the incompleteness of the reference basis set.
    - The f12-exponent $\texttt{gamma}$ is used for the correlation factor $f_{12}=-\texttt{1/gamma}* \texttt{exp(-gamma*}r_{12}\texttt{)}$, which is required for certain, new appearing molecular integrals.
1. Add reference energy and correction to obtain final, corrected energy









```python
# Here, we look at the diatomic Hydrogen molecule.

geomstring = "H 0.0 0.0 0.0\nH 0.0 0.0 {R:.5f}"
distances = [0.5, 0.6, 0.7, 0.8, 0.9, 1, 1.1, 1.2, 1.3, 1.4, 2, 3.000, 4.00 ]
```

## Different choices of CABS basis

### Active space
Within the `active`-approximation, we use a certain basis set for the RI (resolution of the identity), and slice out
an orbital space for the reference calculation by an active space approximation. The remaining basis functions
serve as complementary CABS basis.

Hence, to compute the correction, we need to specify that we want to use `active`-integrals, whereas the reference computation
readily carries specification of the active space.

This strategy works well if one uses a chemically sensible active space and one wants to increase the accuracy. However, results are sensitive to the choice of the active space. Therefore, selecting a large basis set and using some arbitrary orbitals may not lead to accurate energies overall. 

```python
# `Active` procedure for the F12 correction
basis_set = 'cc-pvdz' ##'data/f12correction/sto3g+ccpvdzf12optri'
energy_active = numpy.zeros(len(distances))
corr_active = numpy.zeros(len(distances))
active_orbitals = list(range(4))
# Compute PES of 'active'
for l,r in enumerate(distances):
    geometry = geomstring.format(R=r)
    # Set molecule
    mol = tq.Molecule(geometry=geometry, basis_set=basis_set, active_orbitals=active_orbitals)
    # Compute energy via UCCSD
    U = mol.make_uccsd_ansatz(trotter_steps=1)
    H = mol.make_hamiltonian().simplify()
    E = tq.ExpectationValue(H=H, U=U)
    result = tq.minimize(method='bfgs', objective=E, silent=True, initial_values=0.0)  # HF reference
    energy_active[l] = result.energy

    # Compute RDMs
    mol.compute_rdms(U=U, variables=result.angles)
    # Compute correction
    corr_active[l] = mol.perturbative_f12_correction(rdm1=mol.rdm1, rdm2=mol.rdm2, cabs_type='active', gamma=1.4,
                                                    **{"silent": True})
    
    
print(energy_active, corr_active, energy_active+corr_active)
```

## CABS+
The CABS+ approach [1] builds a complementary space orthogonal to the reference that is exhausted in order to account for the reference's basis set incompleteness. In order to do so, beyond the orbital basis, one needs to provide a complementary auxiliary basis set (CABS). The notion CABS+ was introduced in [1] to denote that the complementary space was constructed using the union of reference and CABS basis, since it was found to be important, that the RI space fully contains the reference.
Note, that this is what we did in the `active` example above, yet without projecting out the reference space from the RI space.

```python
# CABS+ procedure for the F12 correction
basis_set = '6-31g'
energy_cp = numpy.zeros(len(distances))
corr_cp = numpy.zeros(len(distances))
# Compute PES of 'cabs+'
for l,r in enumerate(distances):
    geometry = geomstring.format(R=r)
    # Set molecule
    mol = tq.Molecule(geometry=geometry, basis_set=basis_set, point_group='c1', threads=3)  # CABS+ works only for C1 symmetry

    # Compute RDMs
    mol.compute_rdms(psi4_method='detci', psi4_options={'detci__ex_level': 2}) # CISD is exact here
    energy_cp[l] = mol.logs['detci'].variables["CI TOTAL ENERGY"]
    # Compute correction
    corr_cp[l] = mol.perturbative_f12_correction(rdm1=mol.rdm1, rdm2=mol.rdm2, cabs_type='cabs+', gamma=1.4,
                                              cabs_options={'cabs_name': 'cc-pvdz-f12-optri'}, **{"silent": True})
```

## (MRA-)PNOs as CABS
Last, we look into MRA-PNOs as reference as well as CABS basis, as applied in the context of variation quantum eigensolvers in [2]. See [this tutorial](https://github.com/aspuru-guzik-group/tequila/blob/master/tutorials/ChemistryBasisSetFreeVQE.ipynb)
 for more information on MRA-PNOs. The advantage here is, that PNOs are constructed to be near optimal, which can be expected to be carried over
 to a CABS basis. This means, we potentially require less CABS basis functions to obtain a certain correction (as also found in [1]).
 
The complementary space spanned by PNOs is generated according to the CABS+ procedure [3]. Of course, one could also use a GBS-based CABS with a PNO reference.

Here, we need to provide paths to the externally computed f12-integrals, instead of the f12-exponent.

```python
# Compute some reference using FCI in a QZ-basis
energy_ref = numpy.zeros(len(distances))
for l,r in enumerate(distances):
    geometry = geomstring.format(R=r)
    mol = tq.Molecule(geometry=geometry, basis_set='cc-pvqz')
    energy_ref[l] = mol.compute_energy(method='fci')
```

```python
err_active = energy_active+corr_active - energy_ref
err_cp = energy_cp+corr_cp - energy_ref
err_pno = energy_pno+corr_pno - energy_ref
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(35,15))
ax[0].plot(distances, energy_active, label='Active space reference', color='tab:blue')
ax[0].plot(distances, energy_active+corr_active, '--', label='Corrected active space with virtual space as CABS', color='tab:cyan')
ax[1].plot(distances, energy_cp, label='GBS reference', color='tab:red')
ax[1].plot(distances, energy_cp+corr_cp, '--', label='Corrected GBS with CABS+', color='tab:purple')
ax[2].plot(distances, energy_pno[0,:], label='Cheap PNO reference', color='tab:green')
ax[2].plot(distances, energy_pno[0,:]+corr_pno[0,:], '--', label='Corrected cheap PNOs', color='tab:olive')
ax[2].plot(distances, energy_pno[1,:], label='Good PNO reference', color='tab:orange')
ax[2].plot(distances, energy_pno[1,:]+corr_pno[1,:], '--', label='Corrected good PNOs', color='tab:brown')
for i in range(3):
    ax[i].grid(True)
    ax[i].set_xlabel('Bond Distance H-H in Angstrom')
    ax[i].set_ylabel('Energy in Eh')
    ax[i].legend()
    ax[i].set_ylim((-1.17,-0.95))
print("PES for different approaches mentioned before")
plt.show()

fig2, ax2 = plt.subplots(nrows=1,ncols=3, figsize=(35,15))
ax2[0].plot(distances, corr_active, label='Correction active space', color='tab:cyan')
ax2[1].plot(distances, corr_cp, label='Correction GBS/CABS+', color='tab:purple')
ax2[2].plot(distances, corr_pno[0,:], label='Correction cheap PNOs', color='tab:olive')
ax2[2].plot(distances, corr_pno[1,:], label='Correction good PNOs', color='tab:brown')


ax2[0].plot(distances, err_active, label='Error active space corrected', color='tab:cyan')
ax2[1].plot(distances, err_cp, label='Error GBS/CABS+ corrected', color='tab:purple')
ax2[2].plot(distances, err_pno[0,:], label='Error cheap PNOs corrected', color='tab:olive')
ax2[2].plot(distances, err_pno[1,:], label='Error good PNOs corrected', color='tab:brown')
for i in range(3):
    ax2[i].grid(True)
    ax2[i].set_xlabel('Bond Distance H-H in Angstrom')
    ax2[i].set_ylabel('Error w.r.t. reference in Eh / Correction')
    ax2[i].legend()
    ax2[i].fill_between(distances, -1.6e-3, 1.6e-3, color='tab:green', alpha=.3)
    ax2[i].set_ylim((-.015,.035))
print("Discrepancy with respect to reference computation for different, corrected appraoches.\n"+
      "Note, that separate approaches rely on different references and thus are not directly comparable here. The PNO-references for cheap and good should be somewhat comparable however.")
fig2.show()
```

Some brief comments on the results:
- In general, we can see that the correction vanishes for longer bond distances. This makes sense, because then there is no more electron-electron correlation since one can view the system as two separate Hydrogen atoms.
- Here, the CABS+ approach yields a better overall energy + higher correction despite the better reference energy. \n In general, CABS+ is superior over the `active`-approach. However, similarly to [9], using the correction in an active space computation can increase the overall accuracy at very little cost (and no quantum cost).
- The `good` PNOs perform better than the `cheap` PNOs. Since for the `good` PNOs, there are more orbitals available to make up the complementary space, the correction is higher. Additionally, there is an anomaly for the `cheap` PNOs since using only one macro-iteration when generating the PNOs, no orbitals of the right symmetry have been generated in this case. This could be easily overcome by increasing the number of macro-iterations in the MRA-computation. Larger molecules usually do not suffer from this. [1]
